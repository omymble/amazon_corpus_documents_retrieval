```python
import matplotlib.pyplot as plt

# Data for types
types_data = {
    'subject': 134,
    'author': 32,
    'series': 10,
    'genre': 8,
    'edition': 7,
    'known-item': 7,
    'subject & genre': 7,
    'work': 2,
    'genre & work': 1,
    'subject & author': 1,
    'language': 1,
    'author & genre': 1
}

# Data for genres
genres_data = {
    'literature': 89,
    'history': 60,
    'biography': 24,
    'military': 16,
    'religion': 16,
    'technology': 14,
    'science': 11,
    'education': 8,
    'politics': 4,
    'philosophy': 4,
    'medicine': 3,
    'geography': 3
}

# Extracting labels and values from types data
types_labels = list(types_data.keys())
types_values = list(types_data.values())

# Extracting labels and values from genres data
genres_labels = list(genres_data.keys())
genres_values = list(genres_data.values())

# Create horizontal bar chart for types
plt.figure(figsize=(10, 6))
bars = plt.barh(types_labels, types_values, color='tab:blue', edgecolor='black')
plt.xlabel('Frequency')
plt.title('Distribution of Topic Types')
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, str(int(bar.get_width())), va='center')
plt.tight_layout()
plt.show()

# Create horizontal bar chart for genres
plt.figure(figsize=(10, 6))
bars = plt.barh(genres_labels, genres_values, color='tab:green', edgecolor='black')
plt.xlabel('Frequency')
plt.title('Distribution of Genres')
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, str(int(bar.get_width())), va='center')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# List of categories
categories = [
    'CONTENT#CHARACTERS',
    'CONTENT#PLOT',
    'BOOK#GENERAL',
    'BOOK#AUTHOR',
    'BOOK#GENRE',
    'BOOK#TITLE',
    'BOOK#AUDIENCE',
]

# Confusion matrix (values from the latest diagram)
confusion_matrix = np.array([
    [157, 5, 1, 25, 2, 1, 1],
    [0, 189, 12, 7, 20, 2, 3],
    [2, 1, 90, 10, 2, 1, 10],
    [20, 5, 10, 75, 5, 3, 7],
    [3, 6, 2, 4, 30, 0, 1],
    [1, 0, 0, 3, 1, 35, 0],
    [3, 4, 10, 5, 2, 1, 32],
])

# Plot the confusion matrix
plt.figure(figsize=(7, 6))
ax = sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", cbar=True, xticklabels=categories, yticklabels=categories, linewidths=0.5)

# Customize the plot
ax.set_title("Categories classification", pad=10)
plt.xticks(rotation=45, ha='right')
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()

plt.show()
```
