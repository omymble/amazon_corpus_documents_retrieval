```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split
from pyabsa import AspectTermExtraction as ATEPC

from pyabsa import ATEPCCheckpointManager
# from ATEPC import ATEPCConfigManager, ATEPCTrainer, ABSADatasetList



from config.config import *
```

```python
df = pd.read_pickle(os.path.join(OBTAINED_DATA, 'texts_absa.pkl'))

def convert_to_custom_format(df):
    result = []
    for _, row in df.iterrows():
        for target, polarity in zip(row['targets'], row['polarities']):
            entry = {
                'text': row['text'].replace(target, '$T$'),
                'aspect': target,
                'sentiment': polarity
            }
            result.append(entry)
    return result

# Split the dataset
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
train_df, val_df = train_test_split(train_df, test_size=0.25, random_state=42)  # 0.25 x 0.8 = 0.2

train_data = convert_to_custom_format(train_df)
val_data = convert_to_custom_format(val_df)
test_data = convert_to_custom_format(test_df)

# Save to JSON files
with open('train_absa.json', 'w') as f:
    json.dump(train_data, f, indent=4)
with open('val_absa.json', 'w') as f:
    json.dump(val_data, f, indent=4)
with open('test_absa.json', 'w') as f:
    json.dump(test_data, f, indent=4)

print("Data saved to pyabsa_data.json")
```
